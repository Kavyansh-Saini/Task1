```python
# importing Libraries

import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.colors as colors
%matplotlib inline
```

```python
d=pd.read_csv('sales_data.csv')       #reading CSV file
```

```python
d                   #checking CSV File
```

```python
    # Selecting 1-D Arrays

revenue=d['Revenue']
ID=d['ID']
quantity=d['Quantity']
product=d['Product']
```

```python
# Revenue Over ID  graph

plt.plot(ID,revenue,color='seagreen')
plt.title('Revenue Over ID')
plt.xlabel('ID')
plt.ylabel('Revenue')
plt.grid(linewidth=1, color = 'yellowgreen',)
```

```python
#Revenue and Quantity Scatter Plot

plt.plot(quantity,revenue,'x' ,color='red')
plt.title('Relationship b/w Quantity and Revenue')
plt.xlabel('Quantity')
plt.ylabel('Revenue')
plt.grid(linestyle='--', linewidth=1,color='black')
```

```python
#Product-Revenue Bar graph

plt.bar(product,revenue, width=0.5, color='yellow' ,)
plt.title('Total Revenue for each Product Category')
plt.xlabel('Category')
plt.ylabel('Revelue')
plt.grid(axis='y',linewidth=1,linestyle='--' ,color='black')
```
